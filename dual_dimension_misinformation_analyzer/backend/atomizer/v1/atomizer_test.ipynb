```python
import json
import pandas as pd
from pprint import pprint

from atomizer.atomizer_service import atomize_text
```

```python
def print_atomizer_result(text: str):
    result = atomize_text(text)
    print("=" * 100)
    print("INPUT:")
    print(text)
    print("\nSTATUS:", result["status"])
    print("\nIGNORED SENTENCES:")
    pprint(result["ignored_sentences"])
    print("\nSUMMARY:")
    summary = result["summary"]
    if hasattr(summary, "model_dump"):
        pprint(summary.model_dump())
    else:
        pprint(summary)
    print("\nCLAIM GROUPS:")
    for group in result["claim_groups"]:
        if hasattr(group, "model_dump"):
            group = group.model_dump()
        pprint(group)
        print("-" * 80)
    return result
```

```python
# ============================================================
# Experiment 1: Smoke Test
# ============================================================

sample_text = (
    "I saw this claim in a short online post. "
    "The Great Wall of China is visible from the Moon with the naked eye and the Earth orbits the Sun."
)

smoke_result = print_atomizer_result(sample_text)
```

```python
# ============================================================
# Experiment 2: Subjective / Hearsay Filtering
# ============================================================

subjective_cases = [
    "I think this is fake.",
    "People say coffee cures cancer.",
    "It is claimed that lemon water detoxifies the liver.",
    "The Earth orbits the Sun.",
]

subjective_rows = []

for text in subjective_cases:
    result = atomize_text(text)
    subjective_rows.append({
        "text": text,
        "status": result["status"],
        "ignored_count": len(result["ignored_sentences"]),
        "claim_group_count": len(result["claim_groups"]),
    })

subjective_df = pd.DataFrame(subjective_rows)
display(subjective_df)
```

```python
# ============================================================
# Experiment 3: Conjunction Decomposition
# ============================================================

conjunction_cases = [
    "The Great Wall of China is visible from the Moon with the naked eye and the Earth orbits the Sun.",
    "The Sun and Earth are in the solar system.",
    "Barack Obama was born in Hawaii and served as president of the United States.",
]

for text in conjunction_cases:
    _ = print_atomizer_result(text)
```

```python
# ============================================================
# Experiment 4: Quotation / Reporting Handling
# ============================================================

quotation_cases = [
    'Trump said: "I heard that vaccines cause infertility."',
    'The spokesperson said: "The policy was introduced in 2020."',
    'She said the Earth is flat.',
]

for text in quotation_cases:
    _ = print_atomizer_result(text)
```

```python
# ============================================================
# Experiment 5: Invalid Input Cases
# ============================================================

invalid_cases = [
    "",
    "Hi there",
    "Wow, unbelievable!",
    "I think this is nonsense. People say it is fake.",
]

invalid_rows = []

for text in invalid_cases:
    result = atomize_text(text)
    invalid_rows.append({
        "text": text,
        "status": result["status"],
        "ignored_count": len(result["ignored_sentences"]),
        "claim_group_count": len(result["claim_groups"]),
    })

invalid_df = pd.DataFrame(invalid_rows)
display(invalid_df)
```

```python
# ============================================================
# Experiment 6: Batch Diagnostic Summary
# ============================================================

diagnostic_cases = [
    {"case": "subjective", "text": "I think this is fake."},
    {"case": "single_fact", "text": "The Earth orbits the Sun."},
    {"case": "conjunction", "text": "The Sun and Earth are in the solar system."},
    {"case": "quotation", 'text': 'The spokesperson said: "The policy was introduced in 2020."'},
    {"case": "mixed_long", "text": "I saw this online. China has the largest population in the world and the Earth orbits the Sun."},
]

rows = []

for item in diagnostic_cases:
    result = atomize_text(item["text"])
    summary = result["summary"]
    if hasattr(summary, "model_dump"):
        summary = summary.model_dump()

    rows.append({
        "case": item["case"],
        "status": result["status"],
        "ignored_sentence_count": summary.get("ignored_sentence_count", 0),
        "text_feature_unit_count": summary.get("text_feature_unit_count", 0),
        "fact_check_claim_count": summary.get("fact_check_claim_count", 0),
    })

summary_df = pd.DataFrame(rows)
display(summary_df)
```

```python
# ============================================================
# Experiment 7: Raw JSON Inspection
# ============================================================

inspection_text = 'The Sun and Earth are in the solar system.'
inspection_result = atomize_text(inspection_text)

def to_plain(obj):
    if hasattr(obj, "model_dump"):
        return obj.model_dump()
    if isinstance(obj, list):
        return [to_plain(x) for x in obj]
    if isinstance(obj, dict):
        return {k: to_plain(v) for k, v in obj.items()}
    return obj

print(json.dumps(to_plain(inspection_result), indent=2, ensure_ascii=False))
```

```python
# ============================================================
# Experiment 8: Expected Behavior Check
# ============================================================

expected_cases = [
    {
        "name": "subjective_only",
        "text": "I think this is fake.",
        "expected_status": "invalid_input",
        "expected_claim_groups": 0,
        "expected_fact_claims": 0,
        "expected_ignored_min": 1,
    },
    {
        "name": "single_factual",
        "text": "The Earth orbits the Sun.",
        "expected_status": "success",
        "expected_claim_groups": 1,
        "expected_fact_claims": 1,
        "expected_ignored_min": 0,
    },
    {
        "name": "conjunction_split",
        "text": "The Sun and Earth are in the solar system.",
        "expected_status": "success",
        "expected_claim_groups": 1,
        "expected_fact_claims": 2,
        "expected_ignored_min": 0,
    },
    {
        "name": "mixed_with_subjective_sentence",
        "text": "I heard this online. The Earth orbits the Sun.",
        "expected_status": "success",
        "expected_claim_groups": 1,
        "expected_fact_claims": 1,
        "expected_ignored_min": 1,
    },
    {
        "name": "quotation_case",
        "text": 'The spokesperson said: "The policy was introduced in 2020."',
        "expected_status": "success",
        "expected_claim_groups": 1,
        "expected_fact_claims": 1,
        "expected_ignored_min": 0,
    },
]

check_rows = []

for item in expected_cases:
    result = atomize_text(item["text"])
    summary = result["summary"]
    if hasattr(summary, "model_dump"):
        summary = summary.model_dump()

    actual_status = result["status"]
    actual_claim_groups = len(result["claim_groups"])
    actual_fact_claims = summary.get("fact_check_claim_count", 0)
    actual_ignored = len(result["ignored_sentences"])

    passed = (
        actual_status == item["expected_status"]
        and actual_claim_groups == item["expected_claim_groups"]
        and actual_fact_claims == item["expected_fact_claims"]
        and actual_ignored >= item["expected_ignored_min"]
    )

    check_rows.append({
        "name": item["name"],
        "passed": passed,
        "expected_status": item["expected_status"],
        "actual_status": actual_status,
        "expected_claim_groups": item["expected_claim_groups"],
        "actual_claim_groups": actual_claim_groups,
        "expected_fact_claims": item["expected_fact_claims"],
        "actual_fact_claims": actual_fact_claims,
        "expected_ignored_min": item["expected_ignored_min"],
        "actual_ignored": actual_ignored,
    })

expected_check_df = pd.DataFrame(check_rows)
display(expected_check_df)
```

```python
# Optional: only show failed cases in detail

failed_cases = expected_check_df[expected_check_df["passed"] == False]

if failed_cases.empty:
    print("All expected behavior checks passed.")
else:
    print("Some cases did not match expectations.\n")
    for _, row in failed_cases.iterrows():
        matched_case = next(item for item in expected_cases if item["name"] == row["name"])
        print_atomizer_result(matched_case["text"])
```
