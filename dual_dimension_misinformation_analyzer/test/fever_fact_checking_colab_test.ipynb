```python
# Cell 4: import fact-checking services and choose options
import json
import time
import traceback

import pandas as pd

from api_contract import AnalysisOptions, EachFactualClaimMetadata
from fact_checking.fact_check_service import analyze_fact_check_claims

required_metadata_fields = [
    "retrieval_queries_tried",
    "claim_entities",
    "claim_relation",
    "claim_constraints",
]
missing_metadata_fields = [
    field_name
    for field_name in required_metadata_fields
    if field_name not in EachFactualClaimMetadata.model_fields
]

if missing_metadata_fields:
    print("Metadata schema warning: Colab is importing an older backend schema.")
    print("Missing fields:", missing_metadata_fields)
    print("Restart the runtime, then run from Cell 1 if you want the latest claim-frame audit fields.")
else:
    print("Metadata schema OK: claim-frame audit fields are available.")

options = AnalysisOptions(
    use_query_rewrite=True,
    relevance_threshold=0.1,
    use_oversampling_retry=True,
    use_selective_stabilization=True,
    top_k=3,
    use_all_eligible_evidence=False,
    retrieval_results=8,
)

options.model_dump()
```

```python
# Cell 5: load the FEVER test dataset from Drive
fever_test_path = os.path.join(fever_root, "paper_test.csv")
fever_df = pd.read_csv(fever_test_path)
fever_df["row_index"] = fever_df.index

print("FEVER test path:", fever_test_path)
print("Rows:", len(fever_df))
display(fever_df.head())
display(fever_df["label"].value_counts().rename_axis("label").reset_index(name="count"))
```

```python
# Cell 6: choose a 10-claim balanced FEVER batch
FEVER_RANDOM_SEED = 42
FEVER_LABEL_COUNTS = {
    "SUPPORTS": 4,
    "REFUTES": 3,
    "NOT ENOUGH INFO": 3,
}

sample_rows = []
for label, count in FEVER_LABEL_COUNTS.items():
    label_rows = fever_df[fever_df["label"] == label]
    n = min(count, len(label_rows))
    sample_rows.append(label_rows.sample(n=n, random_state=FEVER_RANDOM_SEED))

fever_sample_df = pd.concat(sample_rows, ignore_index=True)
fever_sample_df = fever_sample_df.sample(frac=1, random_state=FEVER_RANDOM_SEED).reset_index(drop=True)

print("FEVER batch size:", len(fever_sample_df))
display(fever_sample_df[["row_index", "id", "label", "claim"]])
```

```python
# Cell 7: convert FEVER rows into backend claim groups
claim_groups = []
gold_rows = []

for claim_group_id, row in enumerate(fever_sample_df.to_dict(orient="records"), start=1):
    claim_text = str(row["claim"]).strip()
    claim_groups.append({
        "claim_group_id": claim_group_id,
        "original_sentence": claim_text,
        "text_feature_text": claim_text,
        "atomization_applied": False,
        "factual_claims": [
            {
                "fact_claim_id": 1,
                "claim": claim_text,
            }
        ],
    })
    gold_rows.append({
        "claim_group_id": claim_group_id,
        "row_index": int(row["row_index"]),
        "fever_id": int(row["id"]),
        "fever_label": row["label"],
        "claim": claim_text,
    })

gold_df = pd.DataFrame(gold_rows)
display(gold_df)
print(json.dumps(claim_groups, indent=2, ensure_ascii=False))
```

```python
# Cell 8: run FEVER batch one claim at a time
fever_outputs = []
fever_records = []

def build_single_claim_group(row):
    claim_text = str(row["claim"]).strip()
    return [
        {
            "claim_group_id": 1,
            "original_sentence": claim_text,
            "text_feature_text": claim_text,
            "atomization_applied": False,
            "factual_claims": [
                {
                    "fact_claim_id": 1,
                    "claim": claim_text,
                }
            ],
        }
    ]

for run_index, row in fever_sample_df.iterrows():
    row_index = int(row["row_index"])
    claim_text = str(row["claim"]).strip()

    print(f"\n[{run_index + 1}/{len(fever_sample_df)}] row_index={row_index} label={row['label']}")
    print(claim_text)

    started_at = time.time()

    try:
        result = analyze_fact_check_claims(build_single_claim_group(row), options)
        output = result.model_dump()
        fever_outputs.append({
            "row_index": row_index,
            "fever_id": int(row["id"]),
            "fever_label": row["label"],
            "claim": claim_text,
            "output": output,
        })

        if result.factual_claims:
            claim_result = result.factual_claims[0]
            metadata = claim_result.metadata
            metadata_data = metadata.model_dump()

            record = {
                "run_index": run_index + 1,
                "row_index": row_index,
                "fever_id": int(row["id"]),
                "fever_label": row["label"],
                "claim": claim_text,
                "branch_status": result.status,
                "claim_status": claim_result.status,
                "backend_verdict": claim_result.verdict,
                "truth_score": claim_result.truth_score,
                "decision_confidence": claim_result.decision_confidence,
                "evidence_sufficiency": claim_result.evidence_sufficiency,
                "raw_evidence_count": metadata_data.get("search_raw_evidence_count", 0),
                "selected_evidence_count": metadata_data.get("selected_evidence_count", 0),
                "returned_evidence_count": len(claim_result.evidence),
                "retrieval_query_used": metadata_data.get("retrieval_query_used", ""),
                "retrieval_queries_tried": metadata_data.get("retrieval_queries_tried", []),
                "claim_entities": metadata_data.get("claim_entities", []),
                "claim_relation": metadata_data.get("claim_relation", ""),
                "claim_constraints": metadata_data.get("claim_constraints", []),
                "fallback_used": metadata_data.get("fallback_used", False),
                "runtime_seconds": round(time.time() - started_at, 2),
                "error": "",
            }
        else:
            record = {
                "run_index": run_index + 1,
                "row_index": row_index,
                "fever_id": int(row["id"]),
                "fever_label": row["label"],
                "claim": claim_text,
                "branch_status": result.status,
                "claim_status": "missing_claim_result",
                "backend_verdict": None,
                "truth_score": None,
                "decision_confidence": None,
                "evidence_sufficiency": "",
                "raw_evidence_count": None,
                "selected_evidence_count": None,
                "returned_evidence_count": 0,
                "retrieval_query_used": "",
                "retrieval_queries_tried": [],
                "claim_entities": [],
                "claim_relation": "",
                "claim_constraints": [],
                "fallback_used": False,
                "runtime_seconds": round(time.time() - started_at, 2),
                "error": "No factual_claims returned",
            }

    except Exception as error:
        record = {
            "run_index": run_index + 1,
            "row_index": row_index,
            "fever_id": int(row["id"]),
            "fever_label": row["label"],
            "claim": claim_text,
            "branch_status": "exception",
            "claim_status": "exception",
            "backend_verdict": None,
            "truth_score": None,
            "decision_confidence": None,
            "evidence_sufficiency": "",
            "raw_evidence_count": None,
            "selected_evidence_count": None,
            "returned_evidence_count": 0,
            "retrieval_query_used": "",
            "fallback_used": False,
            "runtime_seconds": round(time.time() - started_at, 2),
            "error": repr(error),
        }
        print("ERROR:", repr(error))
        print(traceback.format_exc())

    fever_records.append(record)

fever_result_df = pd.DataFrame(fever_records)
display(fever_result_df)
```

```python
# Cell 9: check FEVER batch stability rules
fever_audit_df = fever_result_df.copy()

fever_audit_df["empty_evidence_but_judged"] = (
    (fever_audit_df["selected_evidence_count"].fillna(0) == 0)
    & (~fever_audit_df["claim_status"].isin(["no_evidence", "invalid_request", "system_error", "exception"]))
)
fever_audit_df["success_schema_incomplete"] = (
    (fever_audit_df["claim_status"] == "success")
    & (
        fever_audit_df["truth_score"].isna()
        | fever_audit_df["backend_verdict"].isna()
        | (fever_audit_df["backend_verdict"] == "")
    )
)
fever_audit_df["no_evidence_with_selected_evidence"] = (
    (fever_audit_df["claim_status"] == "no_evidence")
    & (fever_audit_df["selected_evidence_count"].fillna(0) > 0)
)
fever_audit_df["insufficient_without_evidence"] = (
    (fever_audit_df["claim_status"] == "insufficient_evidence")
    & (fever_audit_df["selected_evidence_count"].fillna(0) == 0)
)
fever_audit_df["system_or_runtime_error"] = fever_audit_df["claim_status"].isin([
    "system_error",
    "exception",
    "missing_claim_result",
])
fever_audit_df["stability_issue"] = (
    fever_audit_df["empty_evidence_but_judged"]
    | fever_audit_df["success_schema_incomplete"]
    | fever_audit_df["no_evidence_with_selected_evidence"]
    | fever_audit_df["insufficient_without_evidence"]
    | fever_audit_df["system_or_runtime_error"]
)

issue_columns = [
    "row_index",
    "fever_id",
    "fever_label",
    "claim_status",
    "backend_verdict",
    "truth_score",
    "evidence_sufficiency",
    "raw_evidence_count",
    "selected_evidence_count",
    "empty_evidence_but_judged",
    "success_schema_incomplete",
    "no_evidence_with_selected_evidence",
    "insufficient_without_evidence",
    "system_or_runtime_error",
    "claim",
    "error",
]

print("Total rows:", len(fever_audit_df))
print("Rows with stability issue:", int(fever_audit_df["stability_issue"].sum()))
display(fever_audit_df[fever_audit_df["stability_issue"]][issue_columns])
```

```python
# Cell 10: calculate FEVER label metrics
FEVER_CLASSES = ["SUPPORTS", "REFUTES", "NOT ENOUGH INFO"]

def backend_verdict_to_fever_label(row):
    verdict = row["backend_verdict"]
    status = row["claim_status"]

    if status in {"no_evidence", "insufficient_evidence"}:
        return "NOT ENOUGH INFO"
    if verdict in {"True", "Mostly True"}:
        return "SUPPORTS"
    if verdict in {"False", "Mostly False"}:
        return "REFUTES"
    if verdict == "Neutral":
        return "NOT ENOUGH INFO"
    return "NOT ENOUGH INFO"

def calculate_macro_f1(df, gold_col, pred_col, classes):
    rows = []
    f1_values = []

    for class_name in classes:
        true_positive = ((df[gold_col] == class_name) & (df[pred_col] == class_name)).sum()
        false_positive = ((df[gold_col] != class_name) & (df[pred_col] == class_name)).sum()
        false_negative = ((df[gold_col] == class_name) & (df[pred_col] != class_name)).sum()

        precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) else 0.0
        recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) else 0.0
        f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0.0

        f1_values.append(f1)
        rows.append({
            "class": class_name,
            "tp": int(true_positive),
            "fp": int(false_positive),
            "fn": int(false_negative),
            "precision": round(precision, 3),
            "recall": round(recall, 3),
            "f1": round(f1, 3),
        })

    return sum(f1_values) / len(f1_values), pd.DataFrame(rows)

fever_metrics_df = fever_audit_df.copy()
fever_metrics_df["predicted_fever_label"] = fever_metrics_df.apply(
    backend_verdict_to_fever_label,
    axis=1,
)
fever_metrics_df["correct"] = fever_metrics_df["predicted_fever_label"] == fever_metrics_df["fever_label"]

macro_f1, per_class_df = calculate_macro_f1(
    fever_metrics_df,
    "fever_label",
    "predicted_fever_label",
    FEVER_CLASSES,
)

summary_metrics_df = pd.DataFrame([
    {"metric": "total_claims", "value": len(fever_metrics_df)},
    {"metric": "accuracy", "value": round(fever_metrics_df["correct"].mean(), 4)},
    {"metric": "macro_f1", "value": round(macro_f1, 4)},
    {"metric": "stability_issues", "value": int(fever_metrics_df["stability_issue"].sum())},
    {"metric": "mean_raw_evidence_count", "value": round(fever_metrics_df["raw_evidence_count"].mean(), 4)},
    {"metric": "mean_selected_evidence_count", "value": round(fever_metrics_df["selected_evidence_count"].mean(), 4)},
    {"metric": "mean_runtime_seconds", "value": round(fever_metrics_df["runtime_seconds"].mean(), 4)},
])

status_summary_df = (
    fever_metrics_df
    .groupby(["fever_label", "claim_status"], dropna=False)
    .size()
    .reset_index(name="count")
    .sort_values(["fever_label", "claim_status"])
)

confusion_df = pd.crosstab(
    fever_metrics_df["fever_label"],
    fever_metrics_df["predicted_fever_label"],
    dropna=False,
)

print("Summary metrics:")
display(summary_metrics_df)

print("Per-class metrics:")
display(per_class_df)

print("Status summary:")
display(status_summary_df)

print("Confusion matrix:")
display(confusion_df)

print("Prediction table:")
display(fever_metrics_df[[
    "row_index",
    "fever_id",
    "fever_label",
    "predicted_fever_label",
    "correct",
    "claim_status",
    "backend_verdict",
    "truth_score",
    "evidence_sufficiency",
    "raw_evidence_count",
    "selected_evidence_count",
    "claim",
]])
```

```python
# Cell 16: calculate FEVER 30 label metrics
fever_30_metrics_df = fever_30_audit_df.copy()

fever_30_metrics_df["predicted_fever_label"] = fever_30_metrics_df.apply(
    backend_verdict_to_fever_label,
    axis=1,
)

fever_30_metrics_df["correct"] = (
    fever_30_metrics_df["predicted_fever_label"] == fever_30_metrics_df["fever_label"]
)

fever_30_macro_f1, fever_30_per_class_df = calculate_macro_f1(
    fever_30_metrics_df,
    "fever_label",
    "predicted_fever_label",
    FEVER_CLASSES,
)

fever_30_summary_metrics_df = pd.DataFrame([
    {"metric": "total_claims", "value": len(fever_30_metrics_df)},
    {"metric": "accuracy", "value": round(fever_30_metrics_df["correct"].mean(), 4)},
    {"metric": "macro_f1", "value": round(fever_30_macro_f1, 4)},
    {"metric": "stability_issues", "value": int(fever_30_metrics_df["stability_issue"].sum())},
    {"metric": "mean_raw_evidence_count", "value": round(fever_30_metrics_df["raw_evidence_count"].mean(), 4)},
    {"metric": "mean_selected_evidence_count", "value": round(fever_30_metrics_df["selected_evidence_count"].mean(), 4)},
    {"metric": "mean_runtime_seconds", "value": round(fever_30_metrics_df["runtime_seconds"].mean(), 4)},
])

fever_30_status_summary_df = (
    fever_30_metrics_df
    .groupby(["fever_label", "claim_status"], dropna=False)
    .size()
    .reset_index(name="count")
    .sort_values(["fever_label", "claim_status"])
)

fever_30_confusion_df = pd.crosstab(
    fever_30_metrics_df["fever_label"],
    fever_30_metrics_df["predicted_fever_label"],
    dropna=False,
)

print("FEVER 30 summary metrics:")
display(fever_30_summary_metrics_df)

print("FEVER 30 per-class metrics:")
display(fever_30_per_class_df)

print("FEVER 30 status summary:")
display(fever_30_status_summary_df)

print("FEVER 30 confusion matrix:")
display(fever_30_confusion_df)

print("FEVER 30 prediction table:")
display(fever_30_metrics_df[[
    "row_index",
    "fever_id",
    "fever_label",
    "predicted_fever_label",
    "correct",
    "claim_status",
    "backend_verdict",
    "truth_score",
    "evidence_sufficiency",
    "raw_evidence_count",
    "selected_evidence_count",
    "claim",
]])
```

```python
# Cell 20: raw retrieval audit for original and final search queries
from fact_checking.gemini_agent import prepare_claim_for_fact_checking
from fact_checking.retrieval_service import choose_evidence, retrieve_evidence

AUDIT_REFUTES_TO_NEI_ROW_INDEX = int(refutes_to_nei_df.iloc[0]["row_index"])

audit_row = fever_df.loc[AUDIT_REFUTES_TO_NEI_ROW_INDEX]
audit_claim = str(audit_row["claim"]).strip()

matching_result_rows = fever_30_result_df[
    fever_30_result_df["row_index"] == AUDIT_REFUTES_TO_NEI_ROW_INDEX
]

if matching_result_rows.empty:
    final_query_used = audit_claim
    batch_queries_tried = []
else:
    matching_result = matching_result_rows.iloc[0]
    final_query_used = str(matching_result["retrieval_query_used"]).strip() or audit_claim
    batch_queries_tried = matching_result.get("retrieval_queries_tried", [])
    if not isinstance(batch_queries_tried, list):
        batch_queries_tried = []

claim_check = prepare_claim_for_fact_checking(
    audit_claim,
    use_query_rewrite=False,
)

print("row_index:", AUDIT_REFUTES_TO_NEI_ROW_INDEX)
print("fever_label:", audit_row["label"])
print("claim:", audit_claim)
print("claim_valid:", claim_check.is_valid_claim)
print("claim_for_verdict:", claim_check.final_claim)
print("final_query_used_in_batch:", final_query_used)
print("queries_tried_in_batch:", batch_queries_tried)

queries_to_audit = []
for query in [audit_claim] + batch_queries_tried + [final_query_used]:
    query = str(query or "").strip()
    if query and query not in [item[1] for item in queries_to_audit]:
        label = "original_claim_query" if query == audit_claim else "batch_query"
        if query == final_query_used and query != audit_claim:
            label = "final_batch_query"
        queries_to_audit.append((label, query))

for query_label, search_query in queries_to_audit:
    print("\n===", query_label, "===")
    print("search_query:", search_query)

    retrieval = retrieve_evidence(
        search_query=search_query,
        retrieval_results=options.retrieval_results,
        use_oversampling_retry=options.use_oversampling_retry,
    )

    raw_rows = []
    for index, evidence_item in enumerate(retrieval.raw_evidence, start=1):
        raw_rows.append({
            "raw_index": index,
            "url": evidence_item.get("url", ""),
            "source_quality": evidence_item.get("source_quality", ""),
            "source_quality_score": evidence_item.get("source_quality_score", ""),
            "content_preview": evidence_item.get("content", "")[:700],
        })

    print("retrieval_query_used:", retrieval.final_claim)
    print("retrieval_error_type:", retrieval.error_type)
    print("retrieval_error_message:", retrieval.error_message)
    print("raw_evidence_count:", retrieval.search_raw_count)
    display(pd.DataFrame(raw_rows))

    if not retrieval.raw_evidence:
        continue

    selection = choose_evidence(
        final_claim=claim_check.final_claim,
        raw_evidence=retrieval.raw_evidence,
        relevance_threshold=options.relevance_threshold,
        top_k=options.top_k,
        use_all_eligible_evidence=options.use_all_eligible_evidence,
    )

    selected_urls = {evidence.url for evidence in selection.selected_evidence}

    scored_rows = []
    for item in selection.filter_debug.get("scored_evidence", []):
        scored_rows.append({
            "selected_for_verdict": item.get("url", "") in selected_urls,
            "filter_reason": item.get("filter_reason", ""),
            "passed_threshold": item.get("passed_threshold", ""),
            "evidence_quality": item.get("evidence_quality", ""),
            "source_quality": item.get("source_quality", ""),
            "relevance_score": item.get("relevance_score", ""),
            "claim_match_score": item.get("claim_match_score", ""),
            "claim_anchor_score": item.get("claim_anchor_score", ""),
            "number_score": item.get("number_score", ""),
            "number_status": item.get("number_status", ""),
            "final_match_score": item.get("final_match_score", ""),
            "selection_priority": item.get("selection_priority", ""),
            "claim_anchors": item.get("claim_anchors", ""),
            "missing_claim_anchors": item.get("missing_claim_anchors", ""),
            "url": item.get("url", ""),
            "content_preview": item.get("content_preview", ""),
        })

    selected_rows = []
    for index, evidence in enumerate(selection.selected_evidence, start=1):
        selected_rows.append({
            "selected_index": index,
            "evidence_quality": evidence.evidence_quality,
            "url": evidence.url,
            "content_preview": evidence.content[:700],
        })

    print("selection_final_claim:", selection.final_claim)
    print("selected_evidence_count:", len(selection.selected_evidence))
    display(pd.DataFrame(scored_rows))
    display(pd.DataFrame(selected_rows))
```
